```python
import csv
import numpy as np
from scipy.optimize import curve_fit
import matplotlib.pyplot as plt
import pandas as pd
import math

freq = [400,800,1200] #MHz
res = [0.65,0.9,1.3] #mOhm
```

```python
def func(x,a,b):
   return a*x**(2/3)+b
popt,pcov = curve_fit(func,freq,res)
print(popt)
plt.scatter(freq,res,marker='.')
x=np.linspace(300,1250)
plt.plot(x,func(x,*popt),color='red')
plt.show()
print(func(7000,*popt))
```
