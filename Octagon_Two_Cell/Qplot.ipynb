```python
import csv
import numpy as np
from scipy.optimize import curve_fit
import matplotlib.pyplot as plt
import pandas as pd

df1 = pd.read_csv("./240417_175209_tracking_Delaminated_Fujikura_Octagon_2-cell_300-6K_without_coupling.csv",encoding = 'utf-8-sig')

t = df1['time']
temp = df1['temp4']
f = df1['Mode1_f0']
Q = df1['Mode1_ql']
```

240319_HTS_First_Measurment

```python
fig,axs = plt.subplots(1,1,figsize=(10,5),dpi=150)
axs.plot(t,temp,marker='.',linestyle='None',markersize=3)
axs.set_xticks(np.arange(0,60001,7500))
axs.set_xticks(np.arange(0,60000,750),minor=True)
axs.set_yticks(np.arange(0,301,50))
axs.set_yticks(np.arange(0,300,5),minor=True)
axs.grid(which='major',alpha=0.7)
axs.grid(which='minor',alpha=0.4)
axs.set_xlabel("Time [s]")
axs.set_ylabel("Temperature [K]")
axs.set_title("Time v. Temperature [300K-10K]")
#plt.close()
```

```python
fig,axs = plt.subplots(1,1,figsize=(10,5),dpi=150)
axs.plot(temp,f,marker='.',linestyle='None',markersize=3)
axs.set_yticks(np.arange(7.051e9,7.080e9,0.005e9))
axs.set_yticks(np.arange(7.050e9,7.080e9,0.001e9),minor=True)
axs.set_xticks(np.arange(0,301,50))
axs.set_xticks(np.arange(0,300,5),minor=True)
axs.grid(which='major',alpha=0.7)
axs.grid(which='minor',alpha=0.4)
axs.set_xlabel("Temperature [K]")
axs.set_ylabel("Frequency [Hz]")
axs.set_title("Temperature v. Frequency [300K-10K]")
#plt.close()
```
